```python
#importing the neccesary library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Loading Dataset with pandas 

df=pd.read_csv("C:\\Users\ABHISHEKXD\Downloads\website_performance.csv")
```

```python
# Check the first few rows of the dataset
print(df.head())

print("________________________________________________________________")
# Get basic statistics of the dataset
print(df.describe())
print("________________________________________________________________")
# Check the data types of each column
print(df.dtypes)
print("________________________________________________________________")
# Check for missing values
print(df.isnull().sum())
print("________________________________________________________________")
```

```python
# Analyze website performance metrics

# Visualize page load time over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Month', y='Page_Load_Time')
plt.xlabel('Month')
plt.ylabel('Page Load Time (ms)')
plt.title('Page Load Time Over Time')
plt.xticks(rotation=45)
plt.show()


```

```python
# Visualize bounce rate over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Month', y='Bounce_Rate')
plt.xlabel('Month')
plt.ylabel('Bounce Rate (%)')
plt.title('Bounce Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python

# Visualize conversion rate over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Month', y='Conversion_Rate')
plt.xlabel('Month')
plt.ylabel('Conversion Rate (%)')
plt.title('Conversion Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```
